```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import datetime as dt
```

```python
df = pd.read_csv("listings.csv")
df = df.dropna()
```

```python
df
```

```python
X = df.drop(['price','id', 'name', 'host_id', 'host_name', 'neighbourhood_group',
             'neighbourhood', 'license','latitude','longitude','calculated_host_listings_count'], axis=1)
```

```python
X
```

```python
y = df['price']
```

```python
y
```

```python
room_hot = pd.get_dummies(X['room_type'], prefix='room')
room_hot
```

```python
X = X.drop('room_type',axis=1).join(room_hot)
```

```python
X
```

```python
X['last_review'] = pd.to_datetime(X['last_review'])
```

```python
X['last_review'] 
```

```python
dateToday = dt.date.today()
```

```python
dateToday
```

```python
todayOrd = dateToday.toordinal()
```

```python
todayOrd
```

```python
X['last_review'] = X['last_review'].apply(dt.datetime.toordinal)
```

```python
X['last_review']
```

```python
X['last_review'] = todayOrd - X['last_review']
```

```python
X['last_review']
```

```python
X.head()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
pipe = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as MSE
import numpy as np
```

```python
np.sqrt(MSE(y_test, y_pred))
```
